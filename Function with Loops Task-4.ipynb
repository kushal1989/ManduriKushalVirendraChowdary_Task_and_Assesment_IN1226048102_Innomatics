```python
def recharge(amount):
    if amount >= 50 and amount in [199, 299, 399, 599]:
        return "Recharge Successful"
    else:
        return "Invalid Recharge"

while True:
    amt = int(input("Enter amount: "))
    print(recharge(amt))
    if recharge(amt) == "Recharge Successful":
        break
```

```python
def check_stock(products):
    for item in products:
        if products[item] < 15:
            print(item, "Reorder Alert")
        else:
            print(item, "Stock OK")

products = {"Rice": 10, "Wheat": 20, "Sugar": 12}
check_stock(products)
```

```python
def result(marks):
    avg = sum(marks) / len(marks)
    if avg >= 50:
        return "Pass"
    else:
        return "Fail"

print(result([45, 60, 55, 40]))
```

```python
def fare(km, peak):
    total = 50 + km * 12
    if peak == "yes":
        total *= 1.25
    return total

while True:
    km = int(input("Enter km: "))
    peak = input("Peak hour (yes/no): ")
    print("Fare:", fare(km, peak))
    if input("Retry? (yes/no): ") == "no":
        break
```

```python
def attendance_check(att):
    p = att.count("P")
    percent = (p / len(att)) * 100
    if percent >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

print(attendance_check(["P","A","P","P","A","P"]))
```

```python
def check_password(pwd):
    if len(pwd) < 8:
        return "Weak"
    digit = False
    special = False
    for ch in pwd:
        if ch.isdigit():
            digit = True
        if ch in "@#$":
            special = True
    if digit and special:
        return "Strong"
    else:
        return "Weak"
print(check_password("abc@1234"))
```
